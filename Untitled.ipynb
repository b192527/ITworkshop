```python
price=df["Price"]
```

```python
import matplotlib.pyplot as plt
```

```python
print(min(price))
print(max(price))
```

```python
fig =plt.figure()
plt.hist(x=price,edgecolor="black",color="grey",bins=55)
plt.xlabel("price range")
plt.ylabel("NO.OF houses")
plt.title("Histogram of prices")
```

```python
rooms_count=df["Rooms"].value_counts()
```

```python
rooms_count=dict(rooms_count)
```

```python
print(rooms_count)
```

```python
a=rooms_count.keys()
b=rooms_count.values()
```

```python
fig=plt.figure(figsize=(6,6))
plt.pie(x=b,labels=a)

plt.legend()
plt.title("pie chart of houses and rooms")
plt.show()
```

```python
#scatter plot
price = df["Price"]
price = list(price)

distance = df["Distance"]
distance = list(distance)
```

```python
fig = plt.figure(figsize = (8,8))

#line plot
plt.plot(distance, price,linestyle="",marker=".",markeredgecolor="purple",markerfacecolor="yellow")
plt.xlabel("Distance")
plt.ylabel("Price")
plt.title("Distance vs Price")
plt.show()
```
